```python
import re
import pandas as pd
import numpy as np
wc = pd.read_csv('wellcome.csv', encoding='ISO8859')

# determine the five most common journals and the total articles for each.
# mean, median, and std of cost/article/journal

# fuzzy wuzzy - fuzzy words library
```

```python
titles = pd.Series(wc["Journal title"])

titles = titles.str.replace("&", "")
titles = titles.str.replace(".", "")
titles = titles.str.capitalize()
titles = titles.str.replace("J ", "Journal ")
titles = titles.str.replace("Chem ", "Chemical")
titles = titles.str.replace("Int ", "International ")
titles = titles.str.replace("biol. ", "biology ")
titles = titles.str.replace("med ", "medicinal ")
titles = titles.replace(r'\bchem\b', 'chemistry', regex=True)
titles = titles.replace(r'\bMol\b', 'Molecular', regex=True)
titles = titles.replace(r'\bclin\b', 'clinical', regex=True)
titles = titles.replace(r'\bbiol\b', 'biology', regex=True)
titles = titles.replace(r'\bj\b', 'journal', regex=True)
titles = titles.replace(r'\bepidemiol\b', 'epidemiology', regex=True)
titles = titles.replace(r'\barchiv\b', 'archive', regex=True)
titles = titles.replace(r'\bjnl\b', 'journal', regex=True)
titles = titles.replace(r'\bbrt\b', 'british', regex=True)
titles = titles.replace('behaviour', 'behavior')
titles = titles.str.strip()
```

```python
wc = wc[['Journal title', 'Article title', 'COST (£) charged to Wellcome (inc VAT when charged)']]
wc = wc.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)':'Cost'})

wc['cleaned_titles'] = titles

wc['Cost'] = wc['Cost'].str.replace('£', '')
wc['Cost'] = wc['Cost'].str.replace('$', '')
wc['Cost'] = wc['Cost'].astype(float)
```

```python
wc = wc[np.abs(wc.Cost-wc.Cost.mean()) <= (3*wc.Cost.std())]
```

```python
statistics = wc.groupby('cleaned_titles').describe()
```

```python
statistics.sort_values([('Cost', 'count')], ascending=False)
```

```python
# created list of unique journal titles to see what titles should be the same 
# but consist of misspellings, use of different characters, or abbreviations

def unique(list1): 
  
    # intilize a null list 
    unique_list = [] 
      
    # traverse for all elements 
    for x in list1: 
        # check if exists in unique_list or not 
        if x not in unique_list: 
            unique_list.append(x) 
        
    # print list 
    for x in unique_list: 
        print(x)

unique(titles)
```
