```python
wdi = pd.read_csv("wdi.csv")
internet = wdi[wdi["Indicator Code"] == "IT.NET.USER.ZS"]
electricity = wdi[wdi["Indicator Code"] == "EG.ELC.ACCS.ZS"]
mobile = wdi[wdi["Indicator Code"] == "IT.CEL.SETS"]
securei = wdi[wdi["Indicator Code"] == "IT.NET.SECR"]
gdp = wdi[wdi["Indicator Code"] == "NY.GDP.PCAP.CD"]

wdi_2 = gdp.append(electricity).append(internet).append(securei).append(mobile)
wdi_2 = pd.DataFrame(wdi_2)

vn_2 = wdi_2[wdi_2["Country Code"] == "VNM"]
ind = wdi_2[wdi_2["Country Code"] == "IND"]
mmr = wdi_2[wdi_2["Country Code"] == "MMR"]
eth = wdi_2[wdi_2["Country Code"] == "ETH"]
syr = wdi_2[wdi_2["Country Code"] == "SYR"]
bgd = wdi_2[wdi_2["Country Code"] == "BGD"]
lka = wdi_2[wdi_2["Country Code"] == "LKA"]
```

```python
for country in vn_2, ind, mmr, eth, syr, bgd, lka:
    country.drop(columns = ['Country Name', 'Country Code', 'Indicator Code'], inplace=True)

```

```python
countries = [vn_2, ind, mmr, eth, syr, bgd, lka]

ctries_rslt = []

for country in countries:
    ctries_rslt.append(country.transpose())
    
for country in ctries_rslt:
    country.rename(columns=country.iloc[0], inplace=True)
    country.drop(country.index[:1], inplace=True)
    country.reset_index(inplace=True)
    country.columns=["Year","GDP","Electricity","Internet","Secure_Internet","Mobile_Subscriptions"]
```

```python
vn = ctries_rslt[0]
ind = ctries_rslt[1]
mmr = ctries_rslt[2]
eth = ctries_rslt[3]
syr = ctries_rslt[4]
bgd = ctries_rslt[5]
lka = ctries_rslt[6]
```

```python
vn = vn.astype(float)
ind = ind.astype(float)
mmr = mmr.astype(float)
eth = eth.astype(float)
syr = syr.astype(float)
bgd = bgd.astype(float)
lka = lka.astype(float)
```

```python
vn['Country'] = "Vietnam"
ind['Country'] = "India"
mmr['Country'] = "Myanmar"
eth['Country'] = "Ethiopia"
syr['Country'] = "Syria"
bgd['Country'] = "Bangladesh"
lka['Country'] = "Sri_Lanka"


allcountries = pd.concat([vn,ind,mmr,eth,syr,bgd,lka])
```

# Choose one variable and plot that variable four different ways.

```python
sns.set()
sns.relplot(x="Year", y="GDP",
            data=ind);
```

```python
sns.set()
sns.relplot(x="Year", y="GDP",
            kind="line",
            data=ind);
```

```python
sns.set()
sns.catplot(x="Year", y="GDP",
            kind="bar",
            data=ind);
```

```python
sns.set()
sns.distplot(ind["GDP"], kde=False, color="b", bins=10)
```

# Choose two continuous variables, and plot them three different ways.

```python
sns.set()
sns.relplot(x="GDP", y="Mobile_Subscriptions", data=ind)
```

```python
sns.set()
sns.relplot(x="GDP", y="Mobile_Subscriptions", hue="Internet", data=ind)
```

```python
sns.set()
sns.relplot(x="GDP", y="Mobile_Subscriptions", size="Internet", sizes=(15,150),data=ind)
```

# Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.catplot(x="Country", y="Internet", kind="swarm",data=allcountries)
```

```python
sns.catplot(x="Country", y="Internet", data=allcountries)
```

```python
sns.catplot(x="Country", y="Internet", kind="box", data=allcountries)
```

```python
sns.catplot(x="Country", y="Internet",
            kind="violin", data=allcountries);
```

```python
sns.catplot(x="Country", y="Internet", kind="bar", data=allcountries);
```

```python
sns.catplot(x="Country", y="Internet", kind="boxen",
            data=allcountries);
```
