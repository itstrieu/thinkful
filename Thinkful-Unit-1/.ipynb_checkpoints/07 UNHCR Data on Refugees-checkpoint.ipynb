```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sm = pd.read_csv('sm_refugees.csv')
sm.set_index('Country or territory of asylum or residence')
sm.columns = ['Country or Territory of Asylum or Residence', 'Year', 'Refugees', 'Refugees Assisted by UNHCR']
smmeans = sm.groupby('Year').aggregate(np.sum)


vn = pd.read_csv('vn_refugees.csv')
vn.set_index('Country or territory of asylum or residence')
vn.columns = ['Country or Territory of Asylum or Residence', 'Year', 'Refugees', 'Refugees Assisted by UNHCR', 'Total Refugees and People in Refugee-Like Situations', "Total Refugees and People in Refugee-Like Situations Assisted by UNHCR"]
vnmeans = vn.groupby('Year').aggregate(np.sum)

plt.figure(figsize=(60, 50))

plt.subplot(5, 5, 1)
plt.plot(vnmeans['Refugees'], color='purple')
plt.rcParams.update({'font.size': 18})
plt.ylabel('Refugees')
plt.xlabel('Year')
plt.title('Refugees from Vietnam 2006-2016')

plt.subplot(5, 5, 2)
plt.plot(vnmeans['Refugees Assisted by UNHCR'], color='green')
plt.rcParams.update({'font.size': 18})
plt.ylabel('Refugees')
plt.xlabel('Year')
plt.title('Vietnamese Refugees Assisted by the UNHCR 2006-2016')
plt.show()

plt.figure(figsize=(60, 50))

plt.subplot(5, 5, 1)
plt.plot(smmeans['Refugees'], color='purple')
plt.rcParams.update({'font.size': 18})
plt.ylabel('Refugees')
plt.xlabel('Year')
plt.title('Refugees from Somalia 2006-2016')

plt.subplot(5, 5, 2)
plt.plot(smmeans['Refugees Assisted by UNHCR'], color='green')
plt.rcParams.update({'font.size': 18})
plt.ylabel('Refugees')
plt.xlabel('Year')
plt.title('Somali Refugees Assisted by UNHCR 2006-2016')
plt.show()
```

**Sum of Vietnamese and Somali Refugees Placed in Other Countries from 2006 to 2016**

The above plots show the number of refugees from Vietnam (top) and Somalia (bottom) who have found asylum in other countries and territories (in purple) and how many of those were assisted by the UNHCR (in green). The top two graphs shows the UNHCR has decreased assistance to Vietnamese asylum seekers from 2006-2007, with assistance remaining steady from 2007-2016. This is despite changes in the number of asylum seekers placed in new countries and territories. One explanation for the large drop in number of Vietnamese refugees is the strengthened stability of the Vietnamese government along with better diplomatic ties with the United States. The sharp increase in refugees in 2008 may be due to a crackdown on dissidence following that period of relative calm. Crackdowns on dissidence come in cycles in Vietnam. The crackdown on dissidence beginning in 2016 may has caused higher numbers of Vietnamese people to seek asylum, as seen in the uptick at the end of the graph. It is expected to have grown up into the present day. 

After examining data on Vietnamese refugees, I chose to look at data on Somalia to see if the kind of crisis refugees face effects the UNHCR's involvement in refugee assistance. In this case, assistance from the UNHCR appears to increase and decrease with the number of total refugees. Additionally, the max sum of Somali refugees is more than triple that of Vietnamese refugees, indicating higher need. Since the government in Somalia collapsed in 1991, Somalia has been viewed as a kind of failed state. Refugees have poured out into surrounding countries as well as been taken in states further away such as the United States. This kind of crisis has likely encouraged or required more assistance from the UNHCR in comparison to the situation in Vietnam, where only those who actively dissent are in immediate danger of persecution. 

```python
au = pd.read_csv('australia_refugees_2016_30.csv')
au.set_index('Country or territory of origin')
au.columns = ['Country or Territory of origin', 'Refugees', 'Refugees Assisted by UNHCR', 'Total Refugees and People in Refugee-Like Situations', "Total Refugees and People in Refugee-Like Situations Assisted by UNHCR"]

gm = pd.read_csv('germany_refugees_2016_30.csv')
gm.set_index('Country or territory of origin')
gm.columns = ['Country or Territory of origin', 'Refugees', 'Refugees Assisted by UNHCR', 'Total Refugees and People in Refugee-Like Situations', "Total Refugees and People in Refugee-Like Situations Assisted by UNHCR"]

us = pd.read_csv('us_refugees_2016_30.csv')
us.set_index('Country or territory of origin')
us.columns = ['Country or Territory of origin', 'Refugees']
```

```python
plt.figure(figsize=(15, 5))
plt.bar(au['Country or Territory of origin'], au['Refugees'])
plt.ylabel('Refugees')
plt.xlabel('Country of Origin')
plt.title('Refugees in Australia')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15, 5))
plt.bar(gm['Country or Territory of origin'], gm['Refugees'], color='purple')
plt.ylabel('Refugees')
plt.xlabel('Country of Origin')
plt.title('Refugees in Germany')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15, 5))
plt.bar(us['Country or Territory of origin'], us['Refugees'], color='green')
plt.ylabel('Refugees')
plt.xlabel('Country of Origin')
plt.title('Refugees in the United States')
plt.xticks(rotation=90)
plt.show()
```

**Refugees settled in Australia, Germany, and the United States in 2016**
